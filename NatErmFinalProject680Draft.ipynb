```python
# to find out the date range of the dataset
df['Dates'] = df['recording_date'].apply(
    lambda v: v if pd.isnull(v) else datetime.datetime.strptime(v, '%Y-%m-%d')
)
print(df['Dates'].min(), df['Dates'].max())
```

```python
#What are top 15 of the most viewed talks?
most_viewed_talks = df[['talk__name', 'view_count', 'recording_date']].sort_values('view_count', ascending=False)[:15]
most_viewed_talks
```

```python
#What are top 15 of the most commented talks?
most_commented_talks = df[['talk__name', 'comment_count', 'recording_date']].sort_values('comment_count', ascending=False)[:15]
most_commented_talks
```
